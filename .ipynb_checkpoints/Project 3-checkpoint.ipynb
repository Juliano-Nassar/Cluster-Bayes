# Pokemon Svm

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Create Database
PokeData = pd.read_csv("Pokemon.csv")

n = int(len(PokeData)/2)
PokeData.head()
```

```python
#Legendary -1  
#not legendary -0
def Legend(i):
    if i:
        return 1
    else:
        return 0
    
PokeData["Legendary"] = PokeData["Legendary"].apply(Legend)
PokeData.head()
```

```python
#Training and testing data
PokeTeste = PokeData[:n]
PokeTreino = PokeData[n:]
```

```python
#Filter only the important columns
StatsTreino = PokeTreino.loc[:,["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]]
StatsTeste = PokeTeste.loc[:,["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]]
StatsTreino.head()
```

```python
#Linear classifier
from sklearn import svm
Classificador = svm.SVC(kernel='linear', C=1)
Classificador.fit(StatsTreino,PokeTreino["Legendary"])
```

```python
#Testing linear classifier
ListaTeste = Classificador.predict(StatsTeste)

#Real values
x = PokeTeste.Legendary

#Comparing values
contador = 0
h = 0
for i in ListaTeste:
    if i == x[h]:
        contador+=1
    h+=1
    
prob = contador/len(x)
print("Prob of getting right: {0}".format(prob))
```

```python
#Exponencial classifier
Classificador = svm.SVC(kernel='rbf',C = 1e6)
Classificador.fit(StatsTreino,PokeTreino["Legendary"])
```

```python
#Testing exponencial classifier
Teste = Classificador.predict(StatsTeste)

#Real values
x = PokeTeste.Legendary

#Comparing values
contador = 0
h = 0

for i in Teste:
    if i == x[h]:
        contador+=1
    h+=1
    
prob = contador/len(x)
print("Prob of getting right: {0}".format(prob))
```

```python
#Formatting type 1
#Water - 1
#Normal - 2
# ...
def Classes(i):
        lista = [
"Water",       
"Normal",       
"Grass" ,       
"Bug" ,         
"Psychic",      
"Fire",         
"Electric",     
"Rock",         
"Ground",       
"Dragon",       
"Ghost",        
"Dark",         
"Poison",       
"Fighting",     
"Steel",        
"Ice",         
"Fairy",       
"Flying" 
        ]
        
        contador = 1
        for h in lista:
            if i == h:
                return contador
            contador+= 1
PokeData["Type 1"] = PokeData["Type 1"].apply(Classes)
PokeData.head()
```

```python
#Training and testing data
PokeTeste = PokeData[:n]
PokeTreino = PokeData[n:]

StatsTreino = PokeTreino.loc[:,["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]]
StatsTeste = PokeTeste.loc[:,["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]]
```

```python
#Multi-Class classifier
Classificador = svm.SVC(decision_function_shape='ovr',kernel='linear')
Classificador.fit(StatsTreino,PokeTreino["Type 1"])
```

```python
#Testing Multi-Class classifier
Teste = Classificador.predict(StatsTeste)

#Real values
x = PokeTeste["Type 1"]

#Comparing values
contador = 0
h = 0

for i in Teste:
    if i == x[h]:
        contador+=1
    h+=1

prob = contador/len(x)
print("Prob of getting right: {0}".format(prob))
```
